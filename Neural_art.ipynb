**deep learning algorithm that blends the style of one image with the content of another. Rather than simply overlaying one image on top of another, the algorithm**

```python
import tensorflow
import tensorflow_hub as hub
import matplotlib.pyplot
import numpy
```

**loading the model**

```python
model = hub.load('https://www.kaggle.com/models/google/arbitrary-image-stylization-v1/TensorFlow1/256/2')
```

**content image will the main image while style image will be super imposed to content image this will cause the content image to become like abstract**

```python
content_image = matplotlib.pyplot.imread('#path of content image')
style_image = matplotlib.pyplot.imread('#path of style iamge')
```

```python
matplotlib.pyplot.imshow(content_image)
```

```python
matplotlib.pyplot.imshow(style_image)
```

**normalizing the image and converting them from integer to float**

```python
content_image=content_image.astype(numpy.float32)/255.0
style_image=style_image.astype(numpy.float32)/255,0
```

```python
content_image.shape,style_image.shape
```

**adding extra dimention to the image .In the format thaat model would  take input**

```python
content_image=tensorflow.expand_dims(content_image,axis=0)
style_image=tensorflow.expand_dims(style_image,axis=0)
```

```python
content_image.shape,style_image.shape
```

```python
image=model(content_image,style_image)
```

```python
matplotlib.pyplot.imshow(image[0][0])
```
